```python
import numpy as np
import pandas as pd
import  matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import LocalOutlierFactor
from sklearn.metrics import classification_report, accuracy_score, mean_absolute_error, mean_squared_error, roc_auc_score, roc_curve, auc, confusion_matrix


pd.options.display.max_columns = None
```

```python
raw_data = pd.read_csv('creditcard.csv')
raw_data['Class'].value_counts()
```

```python
raw_data.isnull().sum()     # No missing values
```

```python
raw_data.describe()
```

```python
raw_data.columns
```

```python
print(f"No frauds: {round(raw_data['Class'].value_counts()[0]/len(raw_data) * 100, 2)}%, {raw_data['Class'].value_counts()[0]}")
print(f"Frauds: {round(raw_data['Class'].value_counts()[1]/len(raw_data) * 100, 2)}%, {raw_data['Class'].value_counts()[1]}")

num_no_fraud = len(raw_data[raw_data['Class'] == 0])
num_fraud = len(raw_data[raw_data['Class'] == 1])
total = num_no_fraud + num_fraud
```

```python
no_fraud_data = raw_data[raw_data['Class'] == 0]
fraud_data = raw_data[raw_data['Class'] == 1]

no_fraud_data = no_fraud_data.sample(fraud_data.shape[0], random_state=365) # Taking equal number of fraud and non-fraud data
no_fraud_data.shape == fraud_data.shape
```

```python
data = pd.concat([no_fraud_data, fraud_data], axis=0) # Concatenating the data to form a single dataset
data
scaler = StandardScaler()

data_without_class = data.drop('Class', axis=1)
scaled_data = scaler.fit_transform(data_without_class)
scaled_data = pd.DataFrame(scaled_data, columns=data_without_class.columns)
scaled_data['Class'] = data['Class'].values
scaled_data.head()
```

```python
# Shuffle the data

scaled_data = scaled_data.sample(frac=1)

scaled_data.reset_index(drop=True, inplace=True) # Resetting the index
```

```python
x = scaled_data.drop('Class', axis=1)
y = scaled_data['Class']
```

```python
outlier_fraction = y.mean()
lof = LocalOutlierFactor(n_neighbors=20, algorithm='auto', 
                         leaf_size=30, metric='minkowski',
                         p=2, contamination=outlier_fraction)
```

```python
#  Fit the model and predict outliers
y_pred_outliers = lof.fit_predict(x)

# LOF outputs -1 for outliers and 1 for inliers, convert these to a boolean mask
mask = y_pred_outliers != -1
```

```python
X_filtered = x[mask]
y_filtered = y[mask]
```

```python
train_x, test_x, train_y, test_y = train_test_split(X_filtered, y_filtered, test_size=0.2, random_state=365)
```

```python
reg = LogisticRegression()

reg.fit(train_x, train_y)
```

```python
reg.score(train_x, train_y)
```

```python
y_train_prob = reg.predict_proba(train_x)[:, 1]

# Compute the AUC score for training set
auc_score_train = roc_auc_score(train_y, y_train_prob)
print('AUC Score (Training):', auc_score_train)
```

```python
y_prediction = reg.predict(test_x)
auc = roc_auc_score(test_y, y_prediction)
```

```python
print(f'Accuracy: {accuracy_score(test_y, y_prediction):.3f}')
print(f'AUC: {auc:.3f}')
```
